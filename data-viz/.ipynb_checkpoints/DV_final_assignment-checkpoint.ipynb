```python
import pandas as pd
import folium
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Topic_Survey_Assignment.csv')
df.set_index('ds_topics', inplace =True)
df.index.rename('', inplace=True)
df
```

```python
df['Very interested'] = [round(x*100/2233, 2) for x in df['Very interested']]
df['Somewhat interested'] = [round(x*100/2233, 2) for x in df['Somewhat interested']]
df['Not interested'] = [round(x*100/2233, 2) for x in df['Not interested']]
df.sort_values('Very interested', inplace=True, ascending =False)

df
```

```python
As for the chart:
use a figure size of (20, 8),
bar width of 0.8,
use color #5cb85c for the Very interested bars, color #5bc0de for the Somewhat interested bars, and color #d9534f for the Not interested bars,
use font size 14 for the bar labels, percentages, and legend,
use font size 16 for the title, and,
display the percentages above the bars as shown above, and remove the left, top, and right borders.
```

```python
#colors = {'vi_perc': '#5cb85c', 'si_perc':'#5bc0de', 'ni_perc':'#d9534f'}
colors = ['#5cb85c', '#5bc0de', '#d9534f']
ax = df.plot(kind='bar', figsize = (20,8), width = 0.8, color=colors, fontsize=14)
#ax.axis('off')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.set_title('Percentage of Respondents\' Interest in Data Science Areas', fontsize=16)
ax.legend(fontsize=14)
ax.tick_params(labelsize=14)

for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14
               )
plt.show()
```
